```python
#Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#Load Instagram data (fix encoding issue)
df = pd.read_csv(
    r"C:\Users\safe\Desktop\multi-domain-data-analysis-portfolio\data\project5_social_media\Instagram data.csv",
    encoding="latin1"
)

df.head()

```

```python
# Check column names
print(df.columns)
```

```python
# Check missing values
df.isnull().sum()
```

```python
# Drop rows with missing values
df = df.dropna()
```

```python
#  Engagement calculation
df['engagement'] = df['Likes'] + df['Comments'] + df['Shares']
df['engagement_rate'] = df['engagement'] / df['Impressions']
```

```python
# Engagement rate distribution
plt.figure()
plt.hist(df['engagement_rate'], bins=20)
plt.title("Engagement Rate Distribution")
plt.xlabel("Engagement Rate")
plt.ylabel("Frequency")
plt.show()
```

```python
#Likes vs Comments
plt.figure()
plt.scatter(df['Likes'], df['Comments'])
plt.xlabel("Likes")
plt.ylabel("Comments")
plt.title("Likes vs Comments Relationship")
plt.show()
```

```python
# Traffic source contribution
traffic_sources = df[['From Home', 'From Hashtags', 'From Explore', 'From Other']].mean()

plt.figure()
traffic_sources.plot(kind='bar')
plt.title("Average Traffic Source Contribution")
plt.ylabel("Average Reach")
plt.show()
```

```python
# Saves vs Engagement
plt.figure()
plt.scatter(df['Saves'], df['engagement_rate'])
plt.xlabel("Saves")
plt.ylabel("Engagement Rate")
plt.title("Saves vs Engagement Rate")
plt.show()
```

## 📊 Key Insights

1. Posts with higher saves show stronger engagement rates
2. Hashtag discovery contributes significantly to impressions
3. Likes alone are not a reliable indicator of engagement
4. Explore page traffic boosts post visibility

## 🎯 Recommendations

1. Optimize posts for saves and shares
2. Use targeted hashtags to improve reach
3. Focus on content that encourages comments
4. Analyze Explore traffic for viral potential
